```python
import pandas as pd
csv_file_path = 'Dataset.csv'

data = pd.read_csv(csv_file_path)
```

```python
#Let's discover our data set
data.head(15)
```

```python
#As we can see: a is an index, b is a description, c is the first name, d is the last name, e is the income,
#f is the debt, and g is the networth or balance.
#Let's rename them to make it more readable

data.columns = ['ID','Description','FirstName','LastName','Income','Debt','NetWorth']
data.head(-15)
```

```python
#Let's merge the two columns into one. It will help us in our last selection to be more visible.
data['fullName']=data['FirstName']+' '+data['LastName']
```

```python
#data.describe()
data.head()
```

```python
sortedData= data.sort_values(by='NetWorth',ascending=False)
sortedData.head()
```

```python
distinctSortedData = sortedData.drop_duplicates(subset='fullName')
distinctSortedData.head()
```

```python
richestPeople = distinctSortedData.head(3)
print(richestPeople[['fullName','NetWorth']])
```
